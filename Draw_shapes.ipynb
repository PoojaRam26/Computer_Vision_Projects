```python
import cv2
import numpy as np
```

```python
img1=np.zeros((512,512,3),np.uint8)
```

```python
 cv2.line(img1,(0,99),(99,0),(255,0,0),2)
```

```python
cv2.rectangle(img1,(0,60),(200,170),(0,255,0),-1)
```

```python
cv2.circle(img1,(60,60),50,(0,0,255),-1)
```

```python
cv2.ellipse(img1,(160,260),(50,20),0,0,360,(127,127,127),2)
```

```python
points=np.array([[80,2],[125,0],[179,0],[230,5],[30,50]],np.int32)
```

```python
points=points.reshape((-1,1,2))
cv2.polylines(img1,[points],True,(255,255,0))
```

```python
text1='Test Text'
cv2.putText(img1,text1,(100,100),cv2.FONT_HERSHEY_SIMPLEX,5,(255,255,0))
```

```python
cv2.imshow("Shapes",img1)
cv2.waitKey(0)
cv2.destroyWindow('Shapes')
```
