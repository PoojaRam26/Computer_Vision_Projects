# Canny Edge Detection

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import feature,data
%matplotlib inline
```

```python
original =data.camera()
plt.imshow(original,cmap='gray')
plt.axis('off')
plt.show()
```

```python
edges1 =feature.canny(original)
plt.imshow(edges1,cmap='gray')
plt.axis('off')
plt.show()
```

```python
edges1 =feature.canny(original,sigma=3)
plt.imshow(edges1,cmap='gray')
plt.axis('off')
plt.show()
```

```python
edges1 =feature.canny(original,sigma=5)
plt.imshow(edges1,cmap='gray')
plt.axis('off')
plt.show()
```
