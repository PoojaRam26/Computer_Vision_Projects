# KMeans Clustering and Quantization

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Image Quantization

```python
#imgpath=(".\image_processing\Advanced Computer Vision\jp.jpg")
img=cv2.imread("jp.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
z=img.reshape((-1,1))
z=np.float32(z)
criteria=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,10,1.0)


k=2
ret,label1,center1=cv2.kmeans(z,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
center1=np.uint8(center1)
res1 = center1[label1.flatten()]
output1 =res1.reshape(img.shape)

k=4
ret,label1,center1=cv2.kmeans(z,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
center1=np.uint8(center1)
res1 = center1[label1.flatten()]
output2 =res1.reshape(img.shape)

k=15
ret,label1,center1=cv2.kmeans(z,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
center1=np.uint8(center1)
res1 = center1[label1.flatten()]
output3 =res1.reshape(img.shape)



output=[img,output1,output2,output3]
titles=["Original","K=2","K=4","K=16"]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

## KMeans on 1D or 2D Data

### One Dimentional Data

```python
x=np.random.randint(25,100,25)
y=np.random.randint(175,255,25)
z=np.hstack((x,y))
z=z.reshape((50,1))
z=np.float32(z)
plt.hist(z,256,[0,256])
plt.show()
```

```python
criteria=(cv2.TERM_CRITERIA_EPS +cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
flags=cv2.KMEANS_RANDOM_CENTERS
compactness,labels,centers=cv2.kmeans(z,2,None,criteria,10,flags)
print(criteria)
print(flags)
```

```python
A=z[labels==0]
B=z[labels==1]
print(A)
print(B)
```

```python
plt.hist(A,256,[0,256],color="g")
plt.hist(B,256,[0,256],color="b")
plt.hist(centers,32,[0,256],color="r")
plt.show()
```

```python
X=np.random.randint(25,50,(25,2))
Y=np.random.randint(60,85,(25,2))
Z=np.vstack((X,Y))
```

```python
Z=np.float32(Z)
criteria=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
ret,label,center=cv2.kmeans(Z,2,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
A=Z[label.ravel()==0]
B=Z[label.ravel()==1]
print(A)
print(B)
```

```python
plt.scatter(A[:,0],A[:,1])
plt.scatter(B[:,0],B[:,1],c='g')
plt.scatter(center[:,0],center[:,1],s=80,c='r',marker='s')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
