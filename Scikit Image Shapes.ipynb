```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from skimage.draw import (line,polygon,circle,circle_perimeter,ellipse,ellipse_perimeter,bezier_curve)
```

### Defining the Image

```python
img =np.zeros((512,512,3),dtype=np.double)
plt.imshow(img)
```

### Line

```python

rr,cc=line(125,125,10,300)
img[rr,cc,1]=1
plt.imshow(img)
```

### Polygon

```python
poly=np.array(((300,300),(480,32),(380,430),(220,590),(300,300)))
rr,cc=polygon(poly[:,0],poly[:,1],img.shape)
img[rr,cc,1]=1
plt.imshow(img)
```

### Circle

```python
rr,cc=circle(250,250,50,img.shape)
img[rr,cc,:]=(1,1,0)
plt.imshow(img)
```

### Ellipse

```python
rr,cc=ellipse(300,300,150,200,img.shape)
img[rr,cc,2]=1
plt.imshow(img)
```

### Circle Perimeter

```python
rr,cc=circle_perimeter(200,300,15)
img[rr,cc,:]=(0.5,0,0)
plt.imshow(img)
```

### Bezier Perimeter

```python
rr,cc=bezier_curve(70,100,10,10,150,100,1)
img[rr,cc,:]=(1,0,0)
plt.imshow(img)
```

### Ellise Perimeter

```python
rr,cc=ellipse_perimeter(120,400,60,20,orientation=math.pi/4.0)
img[rr,cc,:]=(1,0,1)
plt.imshow(img)
```

```python
rr,cc=ellipse_perimeter(120,400,60,20,orientation=math.pi/4.0)
img[rr,cc,:]=(0,0,1)
plt.imshow(img)
```

```python
rr,cc=ellipse_perimeter(120,400,60,20,orientation=math.pi/2.0)
img[rr,cc,:]=(1,1,1)
plt.imshow(img)
```

```python
plt.imshow(img)
plt.title("Shapes")
plt.xticks([])
plt.yticks([])
plt.show()
```
